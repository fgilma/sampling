**Veamos que funciona el algoritmo de reservoir**

Queremos que la probabilidad de cada elemento sea uniforme 1/n

```python
def reservoir(lista,n,k):
    samples =[]
    #asignamos los primeros k elementos
    for i in range(k):
        samples.append(lista[i])
    # y ahora van entrando nuevos elementos
    for j in range(k,n):
        # Generamos un número aleatorio entre los j numeros
        # así estamos cumpliendo que la probabilidad sea 1/j
        index = random.randint(0, j)        
        if index < k:
            samples[index] = lista[j]
    return samples
```

Si n=5 la probabilidad de cada elemento debe ser 1/5=0.2. **Vamos a comprobarlo!!!**

Repetimos el proceso 1000 veces

```python
lista=[1,2,3,4,5]
# m repeticiones
m=1000
lista_samples=[]
for i in range(m):
    # cogemos muestras de 4 elementos y un stream de n
    sample=reservoir(lista,5,4)
    lista_samples.append(sample)
#contamos 
lista=[j for i in lista_samples for j in i]
dict={i:lista.count(i) for i in lista}
dict
```

```python
plt.bar(dict.keys(),dict.values())
plt.xlabel('valores lista')
plt.ylabel('#')
plt.title('# repeticiones')
plt.show()
```

```python
lista=list(dict.values())
np.array(lista)/sum(lista)
```
